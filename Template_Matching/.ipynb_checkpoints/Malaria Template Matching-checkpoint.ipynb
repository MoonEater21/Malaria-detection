Mengimport Library untuk digunakan dalam Image Processing

```python
# Mengimport Library
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.patches import Circle, Rectangle
from skimage import transform
from skimage.io import imread, imshow
from skimage.color import rgb2gray
from skimage.feature import match_template
from skimage.feature import peak_local_max
```

Pada tahap ini, kita mengimport image dalam direktori Lokal

```python
# Import Image
blood = imread('redbloodcells.png')
plt.figure(num=None, figsize=(8, 6))
plt.axis('off')
imshow(blood)
```

Mengubah Image dari RGB ke Grayscale

```python
# Changed Image Color
blood_gray = rgb2gray(blood)
plt.figure(num=None, figsize=(8, 6))
plt.axis('off')
imshow(blood_gray)
```

Memotong Image untuk dijadikan Template Image sesuai koordinat

```python
# Cropping Image
template = blood_gray[155:200, 24:70]
plt.axis('off')
imshow(template);
```

Melihat Map dari Image Grayscale

```python
result_image = match_template(blood_gray, template)
plt.figure(num=None, figsize=(8, 6))
plt.axis('off')
imshow(result_image)
```

Membuat Red marker untuk Mencocokan dengan Image Template

```python
# Red marker untuk Mencocokan dengan Image Template
x, y = np.unravel_index(np.argmax(result_image), result_image.shape)
template_width, template_height = template.shape
rect = plt.Rectangle((y, x), template_height, template_width, color='r', fc='none')
plt.figure(num=None, figsize=(8, 6))
plt.gca().add_patch(rect)
plt.axis('off')
imshow(blood_gray)
```
