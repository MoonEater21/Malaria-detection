Mengimport Library yang dibutuhkan

```python
# Library Import
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Mengimport gambar yang akan diproses

```python
# Image Import
blood_imgRGB = cv2.imread('redbloodcellRGB.png', cv2.IMREAD_UNCHANGED)
blood_img = cv2.imread('redbloodcell.png', cv2.IMREAD_UNCHANGED)
malaria_img = cv2.imread('malaria.png', cv2.IMREAD_UNCHANGED)
```

Membuat variabel untuk hasil (result) serta lokasi berdasarkan value (nilai)

```python
# Result Variabel & Location
result = cv2.matchTemplate(blood_img, malaria_img, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
max_loc
```

Melihat nilai akurasi maksimum dari lokasi yang sudah dibuat

```python
# Maximum Value
max_val
```

Membuat variabel lebar dan tinggi dari kotak penanda

```python
# Variabel w, h
w = malaria_img.shape[1]
h = malaria_img.shape[0]
```

```python
# Rectangle
cv2.rectangle(blood_img, max_loc, (max_loc[0] + w, max_loc[1] + h), (255,0,0), 0)
```

Membuat image processing menjadi lebih mudah dengan metode thresholding biner

```python
# Image Thresholding
threshold = .43
yloc, xloc = np.where(result >= threshold)
```

Membuat kotak penanda sesuai dengan jumlah kecocokan pada template image

```python
# Multiple Rectangle
len(xloc)
for (x,y) in zip (xloc,yloc): cv2.rectangle (blood_img, (x,y), (x + w, y + h), (255,0,0), 0)
```

Plot

```python
# Membuat Plot Image untuk Figur ditempatkannya Image
fig, axes = plt.subplots(ncols=4, figsize=(18, 18))
ax = axes.ravel()

# Membuat Baris dan Kolom
ax[0] = plt.subplot(2, 4, 1)
ax[1] = plt.subplot(2, 4, 2)
ax[2] = plt.subplot(2, 4, 3, sharex=ax[0], sharey=ax[0])
ax[3] = plt.subplot(2, 4, 4, sharex=ax[0], sharey=ax[0])

# Plot di Kolom Pertama (Original)
ax[0].imshow(blood_imgRGB)
ax[0].set_title('Original')
ax[0].axis('off')

# Plot di Kolom Kedua (Malaria)
ax[1].imshow(malaria_img, cmap=plt.cm.gray)
ax[1].set_title('Malaria')
ax[1].axis('off')

# Plot Image di Kolom Ketiga
ax[2].imshow(result, cmap=plt.cm.gray)
ax[2].set_title('Thresholded (Binary)')
ax[2].axis('off')

# Plot Image di Kolom Keempat
ax[3].imshow(blood_img)
ax[3].set_title('Multiple Template Results')
ax[3].axis('off')

# Menampilkan Plot
plt.show()
```
