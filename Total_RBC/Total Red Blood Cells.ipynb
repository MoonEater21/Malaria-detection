#### Berikut Adalah Langkah - Langkahnya

`Mengimport Library yang dibutuhkan`

```python
import cv2 # Image Processing
import matplotlib.pyplot as plt # Visualization
```

`Mengimport Data dengan menggunakan library CV2`

```python
# Read Data
image = cv2.imread('redbloodcell.png', cv2.COLOR_BGR2RGB)
```

`Membuat Image menjadi Grayscale & membuat Blur agar fokus ke Sel Darah Merah`

```python
# Grayscale & Gaussian Blur
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (17, 17), 0.1)
```

`Membuat Threshold Canny agar lebih mudah diproses`

```python
# Thresholded Canny
canny = cv2.Canny(blur, 70, 170, 3)
```

`Threshold Image Diperlebar agar terkesan tajam dan jelas`

```python
# Dilated Canny
dilated = cv2.dilate(canny, (1,1), iterations=1)
```

`Membuat Figure (Tempat Visualisasi dari code sebelumnya) dengan Matplotlib`

```python
# Figure
fig, axes = plt.subplots(ncols=4, figsize=(20, 20))
ax = axes.ravel()

# Membuat Baris dan Kolom
ax[0] = plt.subplot(2, 4, 1)
ax[1] = plt.subplot(2, 4, 2)
ax[2] = plt.subplot(2, 4, 3, sharex=ax[0], sharey=ax[0])
ax[3] = plt.subplot(2, 4, 4, sharex=ax[0], sharey=ax[0])

# First Column
ax[0].imshow(image)
ax[0].set_title('Original')
ax[0].axis('off')

# Second Column
ax[1].imshow(blur, cmap=plt.cm.gray)
ax[1].set_title('Grayscale')
ax[1].axis('off')

# Third Column
ax[2].imshow(canny, cmap=plt.cm.gray)
ax[2].set_title('Thresholded (Canny)')
ax[2].axis('off')

# Fourth Column
ax[3].imshow(dilated, cmap=plt.cm.gray)
ax[3].set_title('Dilated (Diperlebar)')
ax[3].axis('off')

# Show All
plt.show()
```

`Menghitung Jumlah Sel Darah Merah pada Image`

```python
# Total Red Blood Cells
(cnt, hierarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print("Jumlah sel darah merah yaitu sebanyak", len(cnt),"Sel.")
```

`Visualisasi Dari Jumlah Sel Darah Merah`

```python
# Visualization of Red Blood Cells
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (255, 0, 0), 2)
fig,ax = plt.subplots(ncols=1, figsize=(15, 10))
ax.imshow(rgb);
ax.axis("off");
```
